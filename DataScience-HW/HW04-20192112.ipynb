### 문제1
'Gross (Adjusted)' label을 'Adjusted Gross'로 변경하여 movie변수에 assignment하시오.

```
### 답안을 작성하시오. 
movie = movie.relabeled('Gross (Adjusted)','Adjusted Gross')
```

```
movie
```

### 문제2
Adjusted Gross의 최소값을 구하여 min_adjgross 변수에 저장하시오.

```
### 답안을 작성하시오. 
min_adjgross = movie['Adjusted Gross'].min()
```

```
min_adjgross
```

### 문제3
min_adjgross 대비 각 영화의 Adjusted Gross  비율을 구하여 'Ratio'라는 이름의 column을 movie Table에 추가하여 assign하시오. 

```
### 답안을 작성하시오. 
Ratio = movie['Adjusted Gross']/min_adjgross
movie = movie.with_column(
    'Ratio',Ratio
)
```

```
movie.show(3)
```

### 문제4
Paramount 제작 영화만을 출력하시오.

```
### 답안을 작성하시오. 
movie.where('Studio', 'Paramount')
```

### 문제6

### 문제7
2000년대 영화들 중 Ratio가 2.0이상인 영화들을 Ratio가 높은 순으로 출력하시오. 


```
### 답안을 작성하시오. 
movie.where('Year',are.above(2000)).where('Ratio',are.above(2.0)).sort('Ratio', descending=True)
```

### 문제8
a와 b를 입력받아 Year가 a이상이고 Ratio가 b이상인 영화의 편수를 리턴하는 함수인 count_movies 함수를 작성하시오.
(Hint: Table()객체에는 num_rows와 num_columns라는 속성값을 통해 행과 열의 개수를 각각 표현한다.)

예) print(movie.num_rows)


```
### 답안을 작성하시오. 
def count_movies(a, b):
  print((movie.where('Year',are.above(a)).where('Ratio',are.above(b)).sort('Ratio', descending=True)).num_rows)
```

```
count_movies(2000, 2.0)
```
