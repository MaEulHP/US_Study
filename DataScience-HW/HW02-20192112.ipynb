다음은 100명 학생에 관한 정보를 담고 있는 변수들이다. 
 - sname : 학생 이름
 - noclass : 반 번호
 - gender : 성별
 - height : 키
 - weight : 몸무게
 
아래 코드는 수정없이 사용하면 된다.

```
sname=['s'+format(i, '03d') for i in range(100)]
noclass=np.random.randint(1, 6, 100)
gender=['female']*50 + ['male']*50
height=np.random.randint(160, 190, 100)
weight=np.random.randint(50, 80, 100)
```

### 문제1
sname, noclass, gender, height, weight를 column으로 사용하는 DataFrame 변수 ``student``를 생성하시오. 
(index는 자동으로 생성되는 정수)

```
### 답안을 작성하시오. 
student = pd.DataFrame({'sname':sname,'noclass':noclass,'gender':gender,'height':height,'weight':weight})
```

```
### 지우지 마시오.
student.head()  ### 첫 5행을 출력한 결과이다.
```

### 문제2
1반부터 5반까지 담임교사의 성이 각각 Kim, Lee, Park, Choi, Yun이라고 할 때 반번호(1반~5반)와 담임교사의 이름(tname)을 column으로 가지는 DataFrame 변수 ``teacher``를 생성하시오.  (index는 자동으로 생성되는 정수값이다)

```
### 답안을 작성하시오. 
no = pd.Series([1,2,3,4,5])
tname = pd.Series(["Kim", "Lee","Park", "Choi", "Yun"])
teacher = pd.DataFrame({'noclass':no, 'tname':tname})
```

```
### 지우지 마시오.
teacher
```

### 문제3
``student``와 ``teacher``를 합쳐서 ``info`` 라는 이름의 DataFrame변수를 생성하시오. 
 - noclass column이 key값이 된다.

```
### 답안을 작성하시오. 
info = pd.merge(student, teacher)
```

```
### 지우지 마시오.
info.tail() ### 마지막 5행을 보여준다.
```

### 문제4
$ratio=height-weight$ 일 때 새로운 column인 ``ratio``를 ``info`` 에 추가하시오. 

```
### 답안을 작성하시오. 
info['ratio'] = info['height']-info['weight']
```

```
### 지우지 마시오.
info.head()
```

### 문제5
info에서 sname변수가 index로 되도록 변경한 info_sname 변수를 생성하시오.


```
### 답안을 작성하시오. 
info_sname = info.set_index('sname')
```

### 문제6
반별로 height, weight, ratio의 평균값을 구하여 아래 결과와 같이 dataframe 변수 ``info_avg``에 저장하시오.
 - ``info``와 ``info_sname`` 중 어떤 변수를 사용하여도 좋다.

```
### 답안을 작성하시오. 
info_avg = info.groupby('noclass').mean()
```

```
### 지우지 마시오.
info_avg
```

### 문제7
pivot table을 사용하여 반별, 성별  weight 평균값을 구한 결과를 ``info_pivot`` 변수에 저장하시오. 

```
### 답안을 작성하시오. 
info_pivot = info.pivot_table('weight', index = 'noclass', columns='gender')
```

```
### 지우지 마시오.
info_pivot
```

### 문제8
각 반에서 남학생과 여학생 수를 가리키는 DataFrame 변수 ``ct``를 생성하시오.

(단, 이 때 noclass와 gender 순서로 multi-index하며 column이름은 count로 한다)

```
### 답안을 작성하시오. 

cnt=info.groupby(['noclass','gender'])['gender'].count()
ct = pd.DataFrame({'count':cnt})
```

```
### 지우지 마시오.
ct
```
